### Text Cleaning

clean the text for the messages in dataset

============================

    Takes in a string of text, then performs the following:
    1. Remove all punctuation
    2. Handling lower/upper case
    3. Stemming text
    4. Returns a list of the cleaned text


```python
import nltk
from nltk.corpus import stopwords
import string
from nltk.stem import WordNetLemmatizer 
from nltk.tokenize import RegexpTokenizer
from nltk.corpus import wordnet
```

```python
df["Category"] = df["Category"].map({'ham': 0,'spam':1})
```

```python
wordnet_lemmatizer = WordNetLemmatizer()
tokenizer = RegexpTokenizer(r'[a-z]+')
stop_words = set(stopwords.words('english'))
def preprocess(document):
    document = document.lower() # Convert to lowercase
    words = tokenizer.tokenize(document) # Tokenize
    words = [w for w in words if not w in stop_words] # Removing stopwords
    # Lemmatizing
    for pos in [wordnet.NOUN, wordnet.VERB, wordnet.ADJ, wordnet.ADV]:
        words = [wordnet_lemmatizer.lemmatize(x, pos) for x in words]
    return " ".join(words)
```

```python
df['Processed'] = df['Message'].apply(preprocess)

df.head()
```

### TF- IDF
Term Frequency-Inverse Document Frequency

```python
tfidf_transformer = TfidfTransformer().fit(message_bow)
tfidf4 = tfidf_transformer.transform(message_bow)
print(tfidf4)
```

```python
y = df.iloc[:,0].values  
xtr, xts, ytr, yts = train_test_split(tfidf4 ,y,test_size = 0.2,random_state=1115)
len(yts)
```

Now trying to test the model using the **data** that model never seen before.

```python
x= df['Message']
y= df['Category']

xtrain, xtest, ytrain, ytest = train_test_split(x ,y,test_size = 0.2,random_state=1115)
len(yts)

pipeline_SVC = Pipeline([
    ('cv', CountVectorizer(analyzer=preprocess)),  # strings to token integer counts
    ('tfidf', TfidfTransformer()),
    ('classifier', SVC(C= 100, gamma= 0.01, kernel= 'rbf',probability=True )),  # train on TF-IDF vectors with Random Forest
])
pipeline_SVC.fit(xtrain,ytrain)
```

```python
pred= pipeline_SVC.predict(xtest)
print(classification_report(ytest,pred))
```
